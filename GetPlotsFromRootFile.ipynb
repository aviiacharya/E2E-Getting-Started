## I plot 1D histograms here ##

```python
# Preparing a canvas for drawing
canvas = ROOT.TCanvas("canvas", "", 800, 600)

# Directory to save plots
output_dir = "/uscms_data/d3/aacharya/Q_E_ML_TT/"
if not ROOT.gSystem.AccessPathName(output_dir):  # Check if directory exists
    ROOT.gSystem.mkdir(output_dir)

for branch in tree.GetListOfBranches():
    branch_name = branch.GetName()
    leaf = branch.GetLeaf(branch_name)
    if leaf.GetTypeName() in ["Float_t", "Double_t", "float", "double"]:
        # Creating a histogram for float or double type branches
        hist_name = "hist_{}".format(branch_name)
        hist_title = branch_name
        hist = ROOT.TH1F(hist_name, hist_title, 100, leaf.GetMinimum(), leaf.GetMaximum())
        tree.Draw("{}>>{}".format(branch_name, hist_name))
        canvas.SaveAs("{}/{}.png".format(output_dir, branch_name))
    elif leaf.GetTypeName() in ["Int_t", "UInt_t", "int", "unsigned int"]:
        # Creating a histogram for integer type branches
        hist_name = "hist_{}".format(branch_name)
        hist_title = branch_name
        hist = ROOT.TH1I(hist_name, hist_title, 100, leaf.GetMinimum(), leaf.GetMaximum())
        tree.Draw("{}>>{}".format(branch_name, hist_name))
        canvas.SaveAs("{}/{}.png".format(output_dir, branch_name))
    # Add more conditions here for other types if needed

# Cleanup
file.Close()
```

```python
# Open the ROOT file
file = ROOT.TFile.Open('0307C1DA-E49C-AB4B-9179-C70BE232321E.root', 'READ')
tree = file.Get('Events')  # Replace 'Events' with your tree name
```

## 2D histograms here ##

```python

# Preparing a canvas for drawing
canvas = ROOT.TCanvas("canvas", "", 800, 600)

# Directory to save plots
output_dir = "/uscms_data/d3/aacharya/Q_E_ML_TT/"
if ROOT.gSystem.AccessPathName(output_dir):  # Check if directory does NOT exist
    ROOT.gSystem.mkdir(output_dir)

branches = [branch.GetName() for branch in tree.GetListOfBranches()]

# Generate pairs of branches to create 2D histograms
for i, branch_x in enumerate(branches):
    for branch_y in branches[i+1:]:  # Only pair with subsequent branches to avoid duplication
        leaf_x = tree.GetLeaf(branch_x)
        leaf_y = tree.GetLeaf(branch_y)
        if leaf_x.GetTypeName() in ["Float_t", "Double_t", "float", "double"] and leaf_y.GetTypeName() in ["Float_t", "Double_t", "float", "double"]:
            # Creating a 2D histogram for float or double type branches
            hist_name = "hist2D_{}_{}".format(branch_x, branch_y)
            hist_title = "{} vs {}".format(branch_x, branch_y)
            hist = ROOT.TH2F(hist_name, hist_title, 100, leaf_x.GetMinimum(), leaf_x.GetMaximum(), 100, leaf_y.GetMinimum(), leaf_y.GetMaximum())
            tree.Draw("{}:{}>>{}".format(branch_y, branch_x, hist_name), "", "COLZ")
            canvas.SaveAs("{}/{}.png".format(output_dir, hist_name))
        elif leaf_x.GetTypeName() in ["Int_t", "UInt_t", "int", "unsigned int"] and leaf_y.GetTypeName() in ["Int_t", "UInt_t", "int", "unsigned int"]:
            # Creating a 2D histogram for integer type branches
            hist_name = "hist2D_{}_{}".format(branch_x, branch_y)
            hist_title = "{} vs {}".format(branch_x, branch_y)
            hist = ROOT.TH2I(hist_name, hist_title, 100, int(leaf_x.GetMinimum()), int(leaf_x.GetMaximum()), 100, int(leaf_y.GetMinimum()), int(leaf_y.GetMaximum()))
            tree.Draw("{}:{}>>{}".format(branch_y, branch_x, hist_name), "", "COLZ")
            canvas.SaveAs("{}/{}.png".format(output_dir, hist_name))

# Cleanup
file.Close()
```
